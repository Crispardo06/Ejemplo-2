```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Definir la ecuaci√≥n diferencial
def sistema(t, y, coef):
    return -coef * y

# Par√°metros del sistema
params = {
    "coeficiente": 40,  # Valor del coeficiente en la ecuaci√≥n -2y
    "y0": 10,           # Condici√≥n inicial
    "t_inicio": 2,     # Tiempo inicial
    "t_final": 8,      # Tiempo final
    "num_puntos": 50  # N√∫mero de puntos en la simulaci√≥n
}

# Rango de tiempo y evaluaci√≥n
t_eval = np.linspace(params["t_inicio"], params["t_final"], params["num_puntos"])

# Resolver la ecuaci√≥n diferencial
sol = solve_ivp(sistema, [params["t_inicio"], params["t_final"]], [params["y0"]],
                args=(params["coeficiente"],), t_eval=t_eval)

# Mostrar datos de entrada
print("Datos de Entrada:")
print(f"Coeficiente: {params['coeficiente']}")
print(f"Condici√≥n inicial: y0 = {params['y0']}")
print(f"Intervalo de tiempo: [{params['t_inicio']}, {params['t_final']}]\n")

# Mostrar datos de salida (primeros 10 valores)
print("Datos de Salida: (Primeros 10 puntos)")
for i in range(10):
    print(f"Tiempo: {sol.t[i]:.2f} s, y: {sol.y[0, i]:.4f}")

# Color personalizable de la gr√°fica
color_grafica = "red"  # Cambia el color aqu√≠ ('red', 'green', etc.)

# Graficar la soluci√≥n
plt.plot(sol.t, sol.y[0], color=color_grafica, label="y(t)")
plt.title("Soluci√≥n de dy/dt + 2y = 0")
plt.xlabel("Tiempo (s)")
plt.ylabel("y(t)")
plt.legend()
plt.grid()
plt.show()
```
